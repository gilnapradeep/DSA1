# Calculate the sum of first n numbers(recursion)

```python
def sum_first_n(n):
    if n == 1:
        return 1
    return n + sum_first_n(n - 1)
result = sum_first_n(10)
print(result)
```

# Factorial of a number(recursion) 

```python
def factorial(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    return n * factorial(n - 1)
result = factorial(5)
print(result)
```

# Fibonacci series(recursion)

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n - 2)
result = fibonacci(6)
print(result)
```

# Reverse a string(recursion)

```python
def reverse_string(s):
    if len(s) <= 1:
        return s
    return s[-1] +  reverse_string(s[:-1])
s = 'recursion'
result = reverse_string(s)
print(result)
```

# Check given string is palindrome(recursion)

```python
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])
s = 'level'
result = is_palindrome(s)
print(result)
```

# Largest number(recursion)

```python
def largest(n):
    if len(n) == 1:
        return n[0]
    result = largest(n[1:])
    return n[0] if n[0] > result else result
n = [3, 7, 2, 9, 1]
print(largest(n))
```

```python
def reverse_word(s):
    if len(s) <= 1:
        return s
    return s[-1] + reverse_word(s[:-1])
def reverse_each_word(word):
    words = word.split()
    reversed_words = [reverse_word(s) for s in words]
    return ' '.join(reversed_words)
word = "Python is easy to learn"
result = reverse_each_word(word)
print(result)
```

```python
def remove_char(s, char_to_remove):
    if s == "":
        return ""
    first = s[0]
    rest = s[1:]
    if first == char_to_remove:
        return remove_char(rest, char_to_remove)
    else:
        return first + remove_char(rest, char_to_remove)
s = "banana"
result = remove_char(s,'a')
print(result)
```

```python
def b_s_recursion(arr, target, left = 0, right = None):
    if right is None:
        right = len(arr) - 1
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == target:
        return mid
    if target < arr[mid]:
        return b_s_recursion(arr, target, left, mid - 1)
    else:
        return b_s_recursion(arr, target, mid + 1, right)
arr = [1, 3, 5, 7, 9, 11, 13, 15]
print(b_s_recursion(arr,6))
```

```python
def missing_value(arr):
    if not arr:
        return None
    n = len(arr)
    left = 0
    right = n - 1
    while left < right:
        mid = (left + right)//2
        expected = arr[0] + mid
        if arr[mid] == expected:
            left = mid + 1
        else:
            right = mid
    expected_final = arr[0] + left
    if left < n and arr[left] == expected_final:
        return None
    return expected_final
arr = [1, 2, 3, 4, 6, 7, 8, 9, 10]
print(missing_value(arr))
```
